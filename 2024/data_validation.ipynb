```python
import sklearn
import numpy as np
import pandas as pd
import os
import glob
import pandera as pa

from pandera import Check, Column, DataFrameSchema
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))

sets = []
  
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    sets.append(df)
    print()
```

```python
validated_df = pd.DataFrame()
for d in sets:
    columns = d.columns
    print(columns)
    if len(columns) == 2 and columns[1]==' value':
        non_null_schema = DataFrameSchema({
            "time": Column(float, Check(lambda x: x > 0))
        })
        column1_schema = pa.Column(float, name="time")
        column2_schema = pa.Column(float, name=" value")

        df = column1_schema.validate(df)
        validated_df = column2_schema.validate(df)

        validated_df = df.pipe(column1_schema).pipe(column2_schema)
        
        print(validated_df)

        #print(non_null_schema.validate(df))
        
    #for c in columns:
        
```

```python
times = pd.DataFrame()
values = pd.DataFrame()

count = 0



for count in range(len(sets)):
    index = str(count)
    #if(sets[count])
    times[index] = sets[count][sets[count].columns[0]]
    values[index] = sets[count][sets[count].columns[1]]
    
```

```python
values=values.fillna(0)
times=times.fillna(0)
```

```python
scaler=StandardScaler()
```

```python
scaled_values=scaler.fit_transform(values)
scaled_times=scaler.fit_transform(times) 
ratios=[]
```

```python
pca=PCA(n_components=40, svd_solver = 'auto')
pca.fit(scaled_times)
arr=[]
for i in range(pca.n_components_):
    j=i
    arr.append(j+1)
    ratios.append(pca.explained_variance_)
plt.plot(arr, pca.explained_variance_ratio_)
```

```python
transformed_values=pca.fit_transform(scaled_values)
transformed_times=pca.fit_transform(scaled_times)
```

```python
transformed_values
```

```python
inversed_values=pca.inverse_transform(transformed_values)
inversed_times=pca.inverse_transform(transformed_times)
```
