```python
import sklearn
import numpy as np
import pandas as pd
import os
import glob
import pandera as pa

from pandera import Check, Column, DataFrameSchema
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))

sets = []
  
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    sets.append(df)
    print()
```

```python
validated_df = pd.DataFrame()
for d in sets:
    columns = d.columns
    print(columns)
    if len(columns) == 2 and columns[1]==' value':
        non_null_schema = DataFrameSchema({
            "time": Column(float, Check(lambda x: x > 0))
        })
        column1_schema = pa.Column(float, name="time")
        column2_schema = pa.Column(float, name=" value")

        df = column1_schema.validate(df)
        validated_df = column2_schema.validate(df)

        validated_df = df.pipe(column1_schema).pipe(column2_schema)
        
        print(validated_df)

        #print(non_null_schema.validate(df))
        
    #for c in columns:
        
```
